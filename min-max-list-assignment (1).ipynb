## TODO-1 (10 points)

The first step is to define a function `(sumlist L)` that returns the sum of all the elements in the list `L`. E.g.,

    (sumlist '(1 2 3 4)) = 1+2+3+4

> Note: You may assume that the list `L` is a list of natural numbers. (It's OK if you do something else, like a list of rational numbers, but my solution treats `L` as a list of naturals.

> Note: Don't forget to write unit tests for all functions. The unit tests should help you write the code, particularly in writing the defining equations.


```acl2
(defsnapshot from-the-top)
(defsnapshot todo-1)
(definec sumlist (l :nat-list) :nat
    (if (endp l)
        0 
        (+ (first l) (sumlist (rest l)))
    )
)

(check-expect (sumlist '(1 1 1 1)) 4)
(check-expect (sumlist '(0 0 0)) 0)
(check-expect (sumlist '(3 7 1)) 11)
(check-expect (sumlist '(9 9 9 9 9 9 9 9 9)) 81)
(check-expect (sumlist '()) 0)
```

## TODO-3 (10 points)

Now write the function `(minlist L)` that returns the smallest number in the list `L`. E.g.,

    (minlist (list 2 4 6 8)) = 4

If the list `L` does not have any elements, then `minlist` should return `NIL` (to indicate an error).


```acl2
(defsnapshot todo-3)

(definec Mn (x :nat y :optnat) :nat
     (if (equal x y)
         y
            (if (equal y nil)
                x
                (min x y)
            )
     )
)

(definec minlist (L :nat-list) :optnat
    (if (endp L)
        nil
        (Mn (first L) (minlist (rest L)))
    )
)



(check-expect (minlist (list 2 4 6 8)) 2)
(check-expect (minlist (list 1 1 1 1)) 1)
(check-expect (minlist (list 5)) 5)
(check-expect (minlist ()) nil)
(check-expect (minlist (list 8919 11 135 22 19)) 11)








```

## TODO-4 (10 pts)

Now implement `(maxlist L)`, which should be almost identical to `minlist`.


```acl2
(defsnapshot todo-4)


(definec Mx (x :nat y :optnat) :nat
     (if (equal x y)
         y
            (if (equal y nil)
                x
                (max x y)
            )
     )
)

(definec maxlist (L :nat-list) :optnat
    (if (endp L)
        nil
        (Mx (first L) (maxlist (rest L)))
    )
)



(check-expect (maxlist (list 2 4 6 8)) 8)
(check-expect (maxlist (list 1 1 1 1)) 1)
(check-expect (maxlist (list 5)) 5)
(check-expect (maxlist ()) nil)
(check-expect (maxlist (list 8919 11 135 22 19)) 8919)







```


## TODO-5 (10 points)

Here's something we all believe. If L is a list of naturals, then `minlist` returns a number if and only if `maxlist` returns a number. Use `test?` to verify this property.

> Hint: The hypothesis that `L` is a list of naturals is very important.

> Hint: You can use `iff` instead of `implies` to say "if and only if".

> Note: Whenever you use `test?` check the output to see how many different values were checked. If you see something like "0 counterexamples and 4 witnesses" it means it only ran 4 different tests. That's not very convincing. Mine says "0 counterexamples and 1427 witnesses" which is more reassuring. Also, if you see anything other than "0 counterexamples" it means that either you have a bug, or you wrote the property wrong. Fix it :-)

```acl2
(defsnapshot todo-5)

(test? (implies (nat-listp L) (iff (natp (minlist L)) (natp (maxlist L)))))


```

## Aside

You may be feeling a little strange about TODO-6. I said "`L` is a list of naturals", not "`L` is a non-empty list of naturals". So what happens when `L` is empty? In that case, both `minlist` and `maxlist` return `NIL`, so is this true or false:

    (<= NIL NIL)

Your first instinct might be that it's undefined, since `<=` compares numbers and `NIL` is not number. But ACL2 functions **always** return a value, no matter what. So it's not undefined. It's *something*. What ACL2 in most of these cases is to treat each invalid input as some sort of default value. So when it sees `NIL`, it treats that input as if it was `0`. Since `(<= 0 0)` is true, that's also the value of `(<= NIL NIL)`.

So, what do you think is the value of

    (<= NIL 'b)
    (<= 'a 'b)
    (<= 'b 'a)
